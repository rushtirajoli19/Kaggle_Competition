### Ordinal Feature encode
    Below categorical features are ordinal:
    [LandSlope,GarageFinish,PavedDrive,LotShape,LandContour,MasVnrType,ExterQual,
    BsmtQual,BsmtExposure,KitchenQual,MSZoning,ExterCond,BsmtFinType2,GarageCond,HeatingQC]
Note: Using Data description, Manually mapped feature (relative to SalePrice)

```python
for df in [train_df,test_df]:
    landslope_mapping= {'Gtl' : 3,'Unf':2,'Sev':1,'np.nan':0}
    df["LandSlope"]=df["LandSlope"].map(landslope_mapping)

    garagefinish_mapping = {'Fin':3,'Rfn':2,'Unf':1,'np.nan':0}
    df["GarageFinish"]=df["GarageFinish"].map(garagefinish_mapping)

    paveddrive_mapping = {'Y':3,'P':2,'N':1}
    df.PavedDrive = df["PavedDrive"].map(paveddrive_mapping)

    lotshape_mapping = {'Reg':4, 'IR1':3, 'IR2':2, 'IR3':1}
    df.LotShape = df["LotShape"].map(lotshape_mapping)

    LandContour_mapping ={'Lvl':4, 'Bnk':3, 'Low':1, 'HLS':2}
    df.LandContour = df["LandContour"].map(LandContour_mapping)

    MasVnrType_mapping = {'BrkFace':2, 'None':0, 'Stone':1, 'BrkCmn':3}
    df.MasVnrType = df['MasVnrType'].map(MasVnrType_mapping)

    ExterQual_mapping = {'Gd':3, 'TA':2, 'Ex':4, 'Fa':1}
    df.ExterQual = df["ExterQual"].map(ExterQual_mapping)

    BsmtQual_mapping = {'Gd':3, 'TA':2, 'Ex':4,'Fa':1}
    df.BsmtQual = df["BsmtQual"].map(BsmtQual_mapping)

    BsmtExposure_mapping = {'No':0, 'Gd':3, 'Mn':1, 'Av':2}
    df.BsmtExposure = df["BsmtExposure"].map(BsmtExposure_mapping)

    KitchenQual_mapping = {'Gd':3, 'TA':2, 'Ex':4, 'Fa':1}
    df.KitchenQual = df["KitchenQual"].map(KitchenQual_mapping)
    
    MSZoning_mapping = {'RL':2, 'RM':1, 'C (all)':5, 'FV':4, 'RH':3}
    df.MSZoning= df["MSZoning"].map(MSZoning_mapping)

    ExterCond_mapping = {'TA':3, 'Gd':4, 'Fa':2, 'Po':1,'Ex':5}
    df.ExterCond = df["ExterCond"].map(ExterCond_mapping)

    BsmtFinType2_map= {'Unf':1, 'BLQ':4,  'ALQ':5, 'Rec':3, 'LwQ':2, 'GLQ':6}
    df.BsmtFinType2 = df["BsmtFinType2"].map(BsmtFinType2_map)

    GarageCond_mapping ={'TA':3, 'Fa':2,  'Gd':4, 'Po':1, 'Ex':5}
    df.GarageCond = df["GarageCond"].map(GarageCond_mapping)

    HeatingQC_mapping = {'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1}
    df["HeatingQC"] = df["HeatingQC"].map(HeatingQC_mapping)
```

### Nominal Feature Encode
    Performing "ONE-HOT-ENCODE" in all nominal feature can induce Curse of Dimensionality, Here I'm using Feature Hashing for column that has high unique value width and Ordinal Encoding for column has low unique value
   Feature Hashing is useful technique, when categorical feature has high cardinal values it uses "MurMurhash3 function" to reduce width of the column. You can think of hashed outputs as a finite set of b bins such that when hash function is applied on the same values\categories, they get assigned to the same bin (or subset of bins)  based on the hash value. We can pre-define the value of b which becomes the final size of the encoded feature vector for each categorical attribute that we encode

```python
## Below Columns have high cardinal values

print(test_df["Neighborhood"].nunique())
print(test_df["Exterior2nd"].nunique())
```

```python
from sklearn.feature_extraction import FeatureHasher

fh = FeatureHasher(n_features=5, input_type='string')
```

```python
def ordinal_encode(df):
    nominal_feature=list(df.select_dtypes(['object']).columns)
    for i in nominal_feature:
        mapping = {k:i for i,k in enumerate(df[i].sort_values().unique(),1)}
        df[i]=df[i].map(mapping)
```

    Area("BsmtUnfSF") gives FInished Basement Area

    house

```python
for df in [train_df,test_df]:
    df["Finished_basement"]=df["TotalBsmtSF"] - df["BsmtUnfSF"]
    df.drop(labels=["TotalBsmtSF","BsmtUnfSF"],axis=1,inplace=True)
```

```python
for df in [train_df,test_df]:
    df["House_age"]=df["YrSold"]-df["YearRemodAdd"]
    df.drop(labels=["YrSold","YearRemodAdd"],axis=1,inplace=True)
```

### Feature Selection - Boruta Algorithm
    The Boruta Algorithm is used as wrapper method. I'm using XGBoost to rank feature
   
   Boruta algorithm derive its name from a demon in Slavic mythology who lived in pine forests.
   
   To begin, it adds randomness to the given data set by making shuffled copies of all features, known as Shadow Features.
   
   Then, on this extended data set (original attributes plus shadow attributes), it trains any Tree-based Algorithm of our choice (Random Forest, Xgboost, etc.) and applies a feature importance measure such as Mean Decrease Accuracy to evaluate the importance of each feature.
    
   The Boruta Algorithm checks whether a real feature is more important at each iteration. The best of its shadow features and constantly removes features deemed extremely unimportant.
    
   Finally, the Boruta Algorithm terminates either when all features are confirmed or rejected, or when it reaches limit.
    

```python
from boruta import BorutaPy
import xgboost as xgb 
model=xgb.XGBRegressor()
```

```python
## Splitting Independent and Dependent data to train and find best feature

x=train_df.drop("SalePrice",axis=1).values
y=train_df["SalePrice"].values
x_test = test_df.iloc[:,:].values
```

```python
x.shape, y.shape, x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()

x_train_scale=scale.fit_transform(x)
x_test_scale=scale.fit_transform(x_test)
```

```python
## Defining xgboost as model training in Boruta  

feat_selector=BorutaPy(model,n_estimators='auto',verbose=1)
feat_selector.fit(x_train_scale,y)
```

```python
x_train_selected=feat_selector.transform(x_train_scale)
x_test_selected =feat_selector.transform(x_test_scale)
```

```python
print(x_train_selected.shape)
print(x_test_selected.shape)
```

### Training the model with optimal feature

Here, I'm using XGBoost again!! as the model to predict

```python
## Splitting newly selected feature dataframe for Training and Cross-Validation

from sklearn.model_selection import train_test_split
X_train, x_cross_val, Y_train, y_cross_val = train_test_split(x_train_selected,y,train_size=0.8)
```

```python
print(f"80% of Training Data: {X_train.shape}, {Y_train.shape}")
print(f"20% of Cross Val Data: {x_cross_val.shape}, {y_cross_val.shape}")
```

```python
import xgboost as xgb
Reg_model = xgb.XGBRegressor(objective= 'reg:squarederror',seed = 0,max_depth=5,reg_lambda=0.3)
```

```python
Reg_model.fit(X_train,
              Y_train,
              verbose=True,
              eval_metric='rmse',
              early_stopping_rounds=10,eval_set=[(x_cross_val,y_cross_val)])
```

```python
y_cross_pred=Reg_model.predict(x_cross_val)
```

```python
from sklearn.metrics import r2_score
r2_score(y_cross_val,y_cross_pred)
```
